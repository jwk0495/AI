### 시각화

```python
import matplotlib
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 선 그래프
plt.title("선그래프")
plt.plot([1,2,3,4,5],[2,4,6,3,2])  #plot( x, y)
plt.show
```

```python
# 선 그래프
plt.title("선그래프")
plt.plot([1,2,3,4,5],[2,4,6,3,2])
plt.xlabel("일")   # 가로
plt.ylabel("가격", rotation = 0, labelpad = 20) # 세로   rotation = ? : 회전 가능  labelpad = : 간격 벌리기
plt.show
```

```python
# 선 그래프
plt.title("선그래프")
plt.plot([10,20,30,40,50],[2,4,6,3,2],
         c = 'r', lw=4, ls='--',
         marker="o", ms=8, mec='g', mew=3)
plt.xlabel("일")
plt.ylabel("가격", rotation = 0, labelpad = 20)
plt.show
```

```python
plt.title("선그래프", size = 20)
y = [2,4,6,3,8]
x = [10,20,30,40,50]
x_names = ['1일','2일','3일','4일','5일']
plt.plot(x,y)
plt.xticks(x,x_names, rotation=45)  #  rotation= : 이름이 너무 길 때 기울이기
plt.xlabel("일")
plt.ylabel("가격", rotation=0)
plt.show()
```

```python
plt.title("선그래프", size = 20)
y = [2,4,6,3,8]
x = [10,20,30,40,50]
x_names = ['1일','2일','3일','4일','5일']
plt.plot(x,y)
plt.xticks(x,x_names)
plt.grid()          # 격자 무늬
plt.xlabel("일")
plt.ylabel("가격", rotation=0)
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2],  c='r',label="Game1")
plt.plot([2,6,14,17], c='b', label="Game2")
plt.legend()
plt.show()
```

```python
## 그래프를 여러개로 표시
plt.figure(figsize=(5,6))

plt.subplot(2,4,1)
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2],  c='r',label="Game1")

plt.subplot(2,4,4)
plt.plot([2,6,14,17], c='b', label="Game2")
plt.legend()
plt.show()
```

```python
## 그래프를 여러개로 표시
plt.figure(figsize=(5,6))

## subplot(a, b, c) a:행 // b:열 // c:순서
plt.subplot(2,4,1)
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2],  c='r',label="Game1")

plt.subplot(2,4,2)
plt.plot([2,7,5,2],  c='r',label="Game2")

plt.subplot(2,4,4)
plt.plot([2,6,14,17], c='b', label="Game3")
plt.legend()
plt.show()
```
