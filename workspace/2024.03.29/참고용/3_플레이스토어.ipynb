```python
# 1. 모듈 임포트
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver import Keys
```

```python
# 2-1. 드라이버 옵션 설정 및 브라우저 띄우기
chrom_options = webdriver.ChromeOptions()
chrom_options.add_experimental_option('detach', True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options=chrom_options)
```

```python
# 3. 주소 이동
url = "https://play.google.com/store/games?hl=ko-KR"
driver.get(url)
time.sleep(2)
```

```python
# 4. 실제 앱 주소로 이동
app_package = "com.devsisters.CookieRunForKakao"
app_url = f"https://play.google.com/store/apps/details?id={app_package}&hl=ko-KR"
driver.get(app_url)
```

```python
# 5-1. 댓글 수집
comments = driver.find_elements(By.CSS_SELECTOR, ".h3YV2d")
len(comments) # 화면에 보이는 것만 추출이 됩니다. 리뷰 더보기 클릭 필요
```

```python
# 5-2. 리뷰 더보기 클릭
buttons = driver.find_elements(By.CSS_SELECTOR, ".VfPpkd-Bz112c-LgbsSe")
print(len(buttons)) # 여러개 버튼 확인
```

```python
# 5-3. 리뷰 더보기 버튼 선택후 클릭
for btn in buttons:
    if btn.get_attribute("aria-label") == "평가 및 리뷰 자세히 알아보기":
        review_btn = btn
        break
review_btn.click()
```

```python
# 6-1. 리뷰 확인
reviews = driver.find_elements(By.CSS_SELECTOR, "div.h3YV2d")
print(len(reviews)) # 총 23개 밖에 안됨. 스크롤 다운이 필요
```

```python
# 6-2. 스크롤 다운 후 리뷰 확인
page = driver.find_element(By.CLASS_NAME, 'fysCi')
page.click() # 팝업창 클릭
driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
reviews = driver.find_elements(By.CSS_SELECTOR, "div.h3YV2d")
print(len(reviews)) # 리뷰 증가 - 20개씩 증가
time.sleep(2)
```

```python
# 6-3. while 문으로 스크롤 다운 작성
target_review = 200
while True:
    reviews = driver.find_elements(By.CSS_SELECTOR, "div.h3YV2d")
    if len(reviews) < target_review: # 목표 리뷰 숫자보다 작을 경우
        #스크롤다운
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(1)
    else:
        break
```

```python
result_list = []
for review in reviews:
    result_list.append(review.text)
```
