```python
df['state'] = df['state'].replace('긍정적', 1)
df['state'] = df['state'].replace('부정적', 0)
df
```

```python
positive_df = df[df['state'] == 1]
positive_df
```

```python
count_positive = (df['state'] == 1).sum()
count_positive
```

```python
negative_df = df[df['state'] == 0]
negative_df
```

```python
count_negative = (df['state'] == 1).sum()
count_negative
```

```python
import matplotlib
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
labels = ['Positive (1)', 'Negative (0)']
counts = [count_positive, count_negative]

# 선 그래프 그리기
plt.figure(figsize=(8, 5))
plt.plot(labels, counts, marker='o', linestyle=':', color='b')
plt.title('Comparison of Counts for Positive (1) and Negative (0)')
plt.xlabel('State')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(2,4))
plt.bar(['1', '0'], [count_positive, count_negative], color=['blue', 'red'])
plt.title('긍정적/부정적 댓글반응 비교')
plt.xlabel('긍정or부정')
plt.ylabel('댓글 수')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie([count_positive, count_negative], labels=['Positive (1)', 'Negative (0)'], autopct='%1.1f%%', startangle=140, colors=['blue', 'red'])
plt.title('Comparison of Counts for Positive (1) and Negative (0)')
plt.axis('equal')  # 원을 동그랗게 만듭니다.
plt.show()
```

```python
filtered_df_High = df[df['등장 횟수'] >= 3]
filtered_df_High
```

```python
filtered_df_Low = df[df['등장 횟수'] <= 3]
filtered_df_Low
```

```python
import matplotlib
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
plt.figure(figsize=(10, 6))
plt.plot(filtered_df_High['토큰'], filtered_df_High['등장 횟수'], marker='o', color='b', linestyle=':', label = "High")
plt.plot(filtered_df_Low['토큰'], filtered_df_Low['등장 횟수'], marker='o', color='r', linestyle=':', label = "Low")
plt.title('테스트 댓글', size = 20)
plt.xlabel('Token')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # X축 레이블 회전
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# 세로 막대 차트 그리기
plt.figure(figsize=(10, 6))
plt.bar(filtered_df_High['토큰'], filtered_df_High['등장 횟수'], color='r')
plt.title('Token Appearance Frequency')
plt.xlabel('Token')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # X축 레이블 회전
plt.grid(axis='y')  # Y축 그리드 표시
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(filtered_df_High['등장 횟수'], labels=filtered_df_High['토큰'], autopct='%1.1f%%', startangle=140)
plt.title('Token Appearance Frequency', y=1.05, color = 'b')
plt.axis('equal')  # 원을 동그랗게 만듭니다.
plt.show()
```
