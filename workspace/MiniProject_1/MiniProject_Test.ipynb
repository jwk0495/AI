```python
import matplotlib
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
labels = ['Positive (1)', 'Negative (0)']
counts = [count_positive, count_negative]

# ì„  ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(8, 5))
plt.plot(labels, counts, marker='o', linestyle=':', color='b')
plt.title('Comparison of Counts for Positive (1) and Negative (0)')
plt.xlabel('State')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
