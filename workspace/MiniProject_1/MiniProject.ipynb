```python
count_negative = (df2['state'] == 0).sum()
count_negative
```

```python
import matplotlib
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
plt.figure(figsize=(2,4))
plt.bar(['1', '0'], [count_positive, count_negative], color=['blue', 'red'])
plt.title('긍정적/부정적 댓글반응 비교')
plt.xlabel('긍정or부정')
plt.ylabel('댓글 수')
plt.grid(axis='y')
plt.show()
```

```python
# 한 글자 단어를 필터링하여 새로운 DataFrame 생성
filtered_df = df3[df3['Word'].apply(len) > 1]

# Frequency가 400개 이상인 단어 추출
filtered_df_high = filtered_df[filtered_df['Frequency'] >= 400]

# Frequency가 200 이상이고 400 미만인 단어 추출
filtered_df_mid = filtered_df[(filtered_df['Frequency'] >= 200) & (filtered_df['Frequency'] < 400)]

# 막대 색 지정
colors_high = 'salmon'
colors_mid = 'skyblue'

# 막대 그래프 그리기
plt.figure(figsize=(12, 6))

# Frequency가 400개 이상인 단어들의 막대 그래프
plt.bar(filtered_df_high['Word'], filtered_df_high['Frequency'], color=colors_high, label='400회 이상 언급')

# Frequency가 200 이상이고 400 미만인 단어들의 막대 그래프
plt.bar(filtered_df_mid['Word'], filtered_df_mid['Frequency'], color=colors_mid, label='200회 이상 400회 미만 언급')

plt.title('언급된 단어의 빈도', fontsize=20)  # 소제목 크기 조절
plt.xlabel('단어', fontsize=15)
plt.ylabel('빈도수', fontsize=15)
plt.xticks(rotation=45)  # X축 레이블 회전
plt.legend()  # 범례 표시
plt.grid(axis='y')  # Y축 그리드 표시
plt.tight_layout()
plt.show()
```
