```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
# 웹 드라이버 초기화
driver = webdriver.Chrome()
driver.get("https://youtu.be/S1bcy4h1mS8?si=Ozk5G9CZy0krL1um")
```

```python
# 댓글 수집
result = set()

while True:
    # 댓글 요소 찾기
    comments = driver.find_elements(By.CSS_SELECTOR, "#content-text")

    # 댓글 수 출력
    print(len(comments), "개 댓글 수집")

    # 댓글 추가
    for comment in comments:
        result.add(comment.text)

    # 스크롤
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

    # 스크롤 후 로딩 대기
    time.sleep(2)

    # 다음 댓글이 없으면 종료
    if len(comments) >= 1186:  # 원하는 댓글 개수 지정
        break
```

```python
# 대댓글 추가
replies = driver.find_elements(By.CSS_SELECTOR, ".comment-replies #content-text")
for reply in replies:
    result.append(reply.text)
```

```python
# 결과 출력
print("수집된 댓글 수:", len(result))
for idx, comment in enumerate(result, start=0):
    print(f"{idx}. {comment}")
```

```python
import re

# 결과에서 제외할 용어 리스트
excluded_terms = ["!", "그냥", "ㅋㅋㅋㅋ", "진짜", "너무", "정말", "항상", "와", ",", "내", "좀", "수", "근데", "왜", "그" ,"뭐"]

# 필터링된 결과를 저장할 리스트
filtered_result = []

# 정규 표현식 패턴
pattern = re.compile(r'[^가-힣0-9\s]')  # 한글과 숫자를 제외한 문자를 제거하는 패턴

# 결과를 필터링합니다.
for comment in result:
    # 용어 제외
    if any(term in comment for term in excluded_terms):
        continue
    
    # 정규 표현식을 사용하여 문자열에서 한글과 숫자만 남기고 나머지 문자를 제거합니다.
    filtered_comment = re.sub(pattern, '', comment)
    # 공백 제거
    filtered_comment = ' '.join(filtered_comment.split())
    
    # 필터링된 결과를 저장합니다.
    filtered_result.append(filtered_comment)

# 필터링된 결과를 출력합니다.
print("필터링된 결과:")
for idx, comment in enumerate(filtered_result, start=1):
    print(f"{idx}. {comment}")
```

```python
import nltk
from nltk.tokenize import word_tokenize

# NLTK의 다운로드 매니저를 이용하여 필요한 데이터를 다운로드합니다.
nltk.download('punkt')

# 필터링된 결과를 토큰화하여 저장할 리스트
tokenized_filtered_result = []

# 결과를 토큰화합니다.
for comment in filtered_result:
    # 각 댓글을 단어로 토큰화합니다.
    tokens = word_tokenize(comment)
    
    # 토큰화된 결과를 저장합니다.
    tokenized_filtered_result.append(tokens)

# 토큰화된 결과를 출력합니다.
print("토큰화된 결과:")
for idx, tokens in enumerate(tokenized_filtered_result, start=1):
    print(f"{idx}. {tokens}")
```

```python
from collections import Counter

# 토큰화된 결과를 하나의 리스트로 합칩니다.
all_tokens = [token for tokens in tokenized_filtered_result for token in tokens]

# 각 토큰의 등장 횟수를 세어줍니다.
token_counts = Counter(all_tokens)

# 가장 많이 등장한 순서대로 정렬합니다.
sorted_token_counts = sorted(token_counts.items(), key=lambda x: x[1], reverse=True)

# 정렬된 결과를 출력합니다.
print("가장 많이 등장한 토큰 순서:")
for token, count in sorted_token_counts:
    print(f"{token}: {count}번")
```

```python
import csv

# CSV 파일 경로
csv_file_path = "result2.csv"

# CSV 파일에 저장
with open(csv_file_path, "w", newline="", encoding="utf-8-sig") as csvfile:
    writer = csv.writer(csvfile)
    # 헤더 작성
    writer.writerow(["토큰", "등장 횟수"])
    # 결과 작성
    for token, count in sorted_token_counts:
        writer.writerow([token, count])

print(f"CSV 파일이 {csv_file_path}에 저장되었습니다.")
```
