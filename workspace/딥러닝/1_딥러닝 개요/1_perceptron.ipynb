## 1. 퍼셉트론

### 1. AND 게이트 만들어보기

```python
def and_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -0.7
    result = x1 * w1 + x2 * w2 + b
    
    if result <= 0:
        return 0
    else:
        return 1
```

```python
print("(1,1) 인 경우 :", and_gate(1,1))
print("(1,0) 인 경우 :", and_gate(1,0))
```

### 2. OR 게이트 만들어보기

```python
def or_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -0.5
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
print("(1,1) 인 경우 :", or_gate(1,1))
print("(1,0) 인 경우 :", or_gate(1,0))
print("(0,0) 인 경우 :", or_gate(0,0))
```

### 3. 프로젝트 평가 퍼셉트론 만들어보기

```python
def evaluate_project(x1, x2, x3):
    w1 = 0.6
    w2 = 0.3
    w3 = 0.1
    b = 0.3
    result = x1 * w1 + x2 * w2 + x3 * w3 + b

    if result > 8:
        return 1
    else:
        return 0
```

```python
print("통과: ", evaluate_project(8,8,7))
print("다음 기회에: ", evaluate_project(8,7,6))
```

## 2. 활성화 함수

### 1. 시그모이드 함수

```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid 함수')
plt.show()
```

### 2-2. 하이퍼볼탄젠트 함수(Hyperbolic tangent function)

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh 함수')
plt.show()
```

### 2-3. ReLU 함수

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu 함수')
plt.show()
```

### 2-4. Softmax 함수

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax 함수')
plt.show()
```
