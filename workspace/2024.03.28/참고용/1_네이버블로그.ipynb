```python
# 1. 모듈 임포트
import requests
from bs4 import BeautifulSoup
```

```python
# 2. html 요소 얻기
# 유의할 점 - m.으로 시작하는 모바일 주소로 변환
blog_url = "https://m.blog.naver.com/allthetime_/223317346734"
page= requests.get(blog_url)
```

```python
page
```

```python
# 3. BeatifulSoup 으로 파싱
bs_result = BeautifulSoup(page.text, 'html.parser')
bs_result
```

```python
# 4. 요소 추출
result = bs_result.find_all("span", class_="se-fs-")
```

```python
blog_text = ''
for temp in result:
    blog_text += temp.text
blog_text
```

```python
# \u200b 을 띄어쓰기로 변환
blog_text_reuslt = blog_text.replace('\u200b', ' ')
print(blog_text_reuslt)
```
