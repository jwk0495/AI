```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://m.blog.naver.com/hyuni2mail/223397084072' ## 모바일버전 웹(m.) 을 붙여주면 내용을 가져오기 쉽다
response = requests.get(url)
```

```python
bs_result = BeautifulSoup(response.text, "html.parser")
bs_result
```

```python
# 원하는 정보 가져오기
result = bs_result.find_all("span", class_="se-fs-")
result_list = []
for item in result:
    result_list.append(item.text)
print(result_list)
```
